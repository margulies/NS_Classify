## One vs all - One threshold

```python
from IPython.html.widgets import interact, interactive, fixed
from IPython.html import widgets
%load_ext rmagic
```

```python
from neurosynth.base.dataset import Dataset
dataset = Dataset.load("../data/datasets/abs_topics_filt.pkl")
```

```python
from base.classifiers import OnevsallClassifier
from base.classifiers import PairwiseClassifier
import numpy as np
from base import plotting
import base.classifiers.pairwise as pw
import base.classifiers.onevsall as oa
import base.classifiers
%matplotlib inline
import glob
import pandas as pd
from ggplot import *
```

```python
clfs_11 = [OnevsallClassifier.load(clf  + "/classifier.pkl") for clf in glob.glob("../results/ns_11_OvA_RidgeClassifier_DM_roc_*")]
clfs_20 = [OnevsallClassifier.load(clf + "/classifier.pkl") for clf in glob.glob("../results/ns_20_OvA_RidgeClassifier_DM_roc_*")]
clfs_60 = [OnevsallClassifier.load(clf + "/classifier.pkl") for clf in glob.glob("../results/ns_60_OvA_RidgeClassifier_DM_roc_*")]                                                                   
```

```python
plot_score = []
for clfs in [clfs_11, clfs_20, clfs_60]:
    for clf in clfs:
        plot_score.append([clf.mask_num, clf.thresh, 'clf' , clf.class_score.mean(), oa.average_ns(clf)[0]/(oa.average_ns(clf)[1]*1.0), oa.average_ns(clf)[0]])
        plot_score.append([clf.mask_num, clf.thresh, 'dummy', clf.dummy_score.mean(), 0, 0])
```

The reason that some f1 scores are zero is because there are no true or false positives for that class. The classifier is far too weighed towards the other class , resulting in f-score of 0. F- scores are biased if the positive cases are too few - thus switching to ROC AUC

```python
plot_score = pd.DataFrame(plot_score, columns=['n_regions', 'threshold', 'clf', 'roc_auc', 'ratio', 'n' ])
plot_score['n_regions'] = plot_score['n_regions'].apply(str)
```

```python
ggplot(plot_score[plot_score.clf=='clf'], aes('threshold', 'roc_auc', group='n_regions', color='n_regions')) + geom_line() + geom_point()# + facet_grid('n_regions')
```

```python
ggplot(plot_score[plot_score.clf=='clf'], aes('threshold', 'ratio', group='n_regions', color='n_regions')) + geom_line() + geom_point()# + facet_grid('n_regions')
```

```python
ggplot(plot_score[plot_score.clf=='clf'], aes('threshold', 'n', group='n_regions', color='n_regions')) + geom_line() + geom_point()# + facet_grid('n_regions')
```

### Let's try accuracy instead

```python
from sklearn.metrics import accuracy_score
acc_plot = []
for clfs in [clfs_11, clfs_20, clfs_60]:
    for clf in clfs:
        acc_plot.append([clf.mask_num, clf.thresh, 
                         (np.array(base.classifiers.rescore(clf, accuracy_score)) - np.array(base.classifiers.rescore(clf, accuracy_score, dummy=True))).mean()])
```

```python
acc_plot = pd.DataFrame(acc_plot, columns=['n_regions', 'threshold', 'acc'])
acc_plot['n_regions'] = acc_plot['n_regions'].apply(str)
```

```python
ggplot(acc_plot, aes('threshold', 'acc', group='n_regions', color='n_regions')) + geom_line() + geom_point()
```
