# Active or not - hard zero

```python
clfs_wu = [OnevsallClassifier.load(clf + "/classifier.pkl") for clf in glob.glob("../results/wash_u_200v_OvA_RidgeClassifier_DM_hard0_roc_abs_topics_filt_tn_*")]
```

```python
plot_score = []
all_scores = []
for clfs in [clfs_wu]:
    for clf in clfs:
        for n, score in enumerate(clf.class_score):
            all_scores.append([clf.mask_num, n, clf.thresh, 'clf' , score, oa.average_ns(clf, mean=False)[n][0]/(oa.average_ns(clf,mean=False)[n][1]*1.0), oa.average_ns(clf, mean=False)[n][0]])
        plot_score.append([clf.mask_num, clf.thresh, 'clf' , clf.class_score.mean(), oa.average_ns(clf)[0]/(oa.average_ns(clf)[1]*1.0), oa.average_ns(clf)[0]])
        plot_score.append([clf.mask_num, clf.thresh, 'dummy', clf.dummy_score.mean(), 0, 0])
```

The reason that some f1 scores are zero is because there are no true or false positives for that class. The classifier is far too weighed towards the other class , resulting in f-score of 0. F- scores are biased if the positive cases are too few - thus switching to ROC AUC

```python
plot_score = pd.DataFrame(plot_score, columns=['n_regions', 'threshold', 'clf', 'roc_auc', 'ratio', 'n' ])
plot_score['n_regions'] = plot_score['n_regions'].apply(str)

all_scores = pd.DataFrame(all_scores, columns=['n_regions', 'region', 'threshold', 'clf', 'roc_auc', 'ratio', 'n' ])
all_scores['n_regions'] = all_scores['n_regions'].apply(str)
```

### Is classification accuracy predicted by within / between neighborhood membership?

```python
key = pd.read_csv("../masks/wash_u/parcels_continuous_125_key.csv")
```

```python
communities = pd.read_csv("../masks/wash_u/ParcelCommunities.txt", delimiter=" ", header=None)
```

```python
communities.columns = ['original', 'community']
```

```python
communities = pd.merge(communities, key)
```

```python
cs = tools.copy_along_diagonal(p_clf.class_score)
```

```python
communities[communities['community'] == communities.ix[0].community]['new']
```

```python
communities.shape
```

```python
li = list(communities[communities['community'] == communities.ix[0].community]['new'])
```

```python
cs[0][li]
```

```python
within_community = []
between_community = []
for i, region in enumerate(cs):
    within_community.append(region[list(communities[communities['community'] == communities.ix[i].community]['new'])].mean())
    between_community.append(region[list(communities[communities['community'] != communities.ix[i].community]['new'])].mean())
    
```

```python
between_community = np.array(between_community)
between_community.mean()
```

```python
within_community = np.ma.array(within_community, mask= np.isnan(within_community))
within_community.mean()
```

```python
community_analysis = pd.DataFrame(np.array([between_community, within_community]).T, columns=['between_comm', 'within_comm'])
```

```python
community_analysis['region'] = range(0, community_analysis.shape[0])
```

```python
community_analysis = community_analysis.dropna()
```

```python
pylab.rcParams['figure.figsize'] = 10, 10
sns.violinplot(community_analysis.ix[:, 0:2], inner="point",  join_rm=True, color="husl", )
```

```python
comm_sorted = communities.sort('community')
## Remove community #8
comm_sorted = comm_sorted[comm_sorted['community'] != 8]

lines_comm = list(comm_sorted.groupby('community').aggregate(len)['new'].cumsum())
cs_sorted = cs[comm_sorted['new']][:, comm_sorted['new']]
regions_labels = list(comm_sorted['new'])
community_labels = list(comm_sorted['community'].astype('int'))
plotting.heat_map(cs_sorted, regions_labels, regions_labels, size=(25, 25), lines=lines_comm, lines_off = 0)
```
