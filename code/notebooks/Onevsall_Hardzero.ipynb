```python
clfs_11 = [OnevsallClassifier.load(clf  + "/classifier.pkl") for clf in glob.glob("../results/ns_11_OvA_RidgeClassifier_DM_hard0*")]
clfs_20 = [OnevsallClassifier.load(clf + "/classifier.pkl") for clf in glob.glob("../results/ns_20_OvA_RidgeClassifier_DM_hard0*")]
clfs_30 = [OnevsallClassifier.load(clf + "/classifier.pkl") for clf in glob.glob("../results/ns_30_15_OvA*")]
clfs_60 = [OnevsallClassifier.load(clf + "/classifier.pkl") for clf in glob.glob("../results/ns_60_OvA_RidgeClassifier_DM_hard0*")]                                                                   
```

```python
plot_score = []
all_scores = []
for clfs in [clfs_11, clfs_20, clfs_30, clfs_60]:
    for clf in clfs:
        for n, score in enumerate(clf.class_score):
            all_scores.append([clf.mask_num, n, clf.thresh, 'clf' , score, oa.average_ns(clf, mean=False)[n][0]/(oa.average_ns(clf,mean=False)[n][1]*1.0), oa.average_ns(clf, mean=False)[n][0]])
        plot_score.append([clf.mask_num, clf.thresh, 'clf' , clf.class_score.mean(), oa.average_ns(clf)[0]/(oa.average_ns(clf)[1]*1.0), oa.average_ns(clf)[0]])
        plot_score.append([clf.mask_num, clf.thresh, 'dummy', clf.dummy_score.mean(), 0, 0])
```

The reason that some f1 scores are zero is because there are no true or false positives for that class. The classifier is far too weighed towards the other class , resulting in f-score of 0. F- scores are biased if the positive cases are too few - thus switching to ROC AUC

```python
plot_score = pd.DataFrame(plot_score, columns=['n_regions', 'threshold', 'clf', 'roc_auc', 'ratio', 'n' ])
plot_score['n_regions'] = plot_score['n_regions'].apply(str)

all_scores = pd.DataFrame(all_scores, columns=['n_regions', 'region', 'threshold', 'clf', 'roc_auc', 'ratio', 'n' ])
all_scores['n_regions'] = all_scores['n_regions'].apply(str)
```
