The goal here is to see if adjusting class_weight will render a balanced accuracy score

```python
from sklearn.datasets import make_classification
from sklearn.cross_validation import cross_val_score
import numpy as np
```

```python
from sklearn.linear_model import RidgeClassifier
# Classifier w/ class_weight = None
clf = RidgeClassifier()

# Classifier w/ class_weight set to the reverse of the class frequency
# To be clear, this will train the classifier equally on both classes by subsampling or resampling classes
clf_auto = RidgeClassifier(class_weight={0: 0.7, 1: 0.3})
```

```python
from sklearn.dummy import DummyClassifier
# Three types of dummy classifiers:

# Classifies using class distribution but is probabilistic in nature
dummy_s = DummyClassifier(strategy='stratified')

# Always chooses most frequent
dummy_f = DummyClassifier(strategy='most_frequent')

# Chooses a completely random class
dummy_u = DummyClassifier(strategy='uniform')
```

Below I'm going to make a sample dataset 1000 times. I'm going to test each time using each classifier. 

```python
res_diff = []
res_auto = []
r_n_sum = []
r_a_sum = []
d_s_sum = []
d_f_sum = []
d_u_sum = []

for i in xrange(0, 1000):
    X, y = make_classification(n_samples=200, weights=[0.3, 0.7])
    print "Ridge none:"
    r_n = cross_val_score(clf, X, y).mean()
    print r_n
    r_n_sum.append(r_n)
    
    print "Ridge reverse frequency:"
    r_a = cross_val_score(clf_auto, X, y).mean()
    print r_a
    r_a_sum.append(r_a)
    
    print "Dummy stratified:"
    d_s = cross_val_score(dummy_s, X, y).mean()
    print d_s
    d_s_sum.append(d_s)
    
    print "Dummy frequency"
    d_f  = cross_val_score(dummy_f, X, y).mean()
    print d_f
    d_f_sum.append(d_f)
    
    print "Random dummy"
    d_u = cross_val_score(dummy_u, X, y).mean()
    print d_u
    d_u_sum.append(d_u)
    
    print "Difference between dummy stratified and none"
    print r_n-d_s
    
    res_diff.append(r_n-d_s)
    
    print "Difference between class_weight and 50%"
    print r_a - 0.5
    
    res_auto.append(r_a - 0.5)

res_diff = np.array(res_diff)
res_auto = np.array(res_auto)
r_n_sum = np.array(r_n_sum)
r_a_sum = np.array(r_a_sum)
d_s_sum = np.array(d_s_sum)
d_f_sum = np.array(d_f_sum)
d_u_sum = np.array(d_u_sum)
```

```python
print
print "Means over 1000 iterations:"
print "Ridge none"
print r_n_sum.mean()
print "Ridge reverse frequency"
print r_a_sum.mean()
print "Dummy stratified"
print d_s_sum.mean()
print "Dummy frequency"
print d_f_sum.mean()
print "Dummy random"
print d_u_sum.mean()
print "Difference between dummy stratified and none"
print res_diff.mean()
print "Difference between class_weight and 50%"
print res_auto.mean()

print "Average difference between the two methods (paired)"
print (res_diff - res_auto).mean()
```
