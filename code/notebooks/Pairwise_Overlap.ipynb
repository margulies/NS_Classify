# Pairwise analysis

```python
p_11 = [PairwiseClassifier.load(clf  + "/classifier.pkl") for clf in glob.glob("../results/ns_11_Pairwise_RidgeClassifier_roc_overlap_*/")]
p_20 = [PairwiseClassifier.load(clf + "/classifier.pkl") for clf in glob.glob("../results/ns_20_Pairwise_RidgeClassifier_roc_overlap_*/")]
# p_60 = [PairwiseClassifier.load(clf + "/classifier.pkl") for clf in glob.glob("../results/ns_60_P")]                                                                   
```

```python
pair_score = []
for clfs in [p_11, p_20]:
    for clf in clfs:
        pair_score.append([clf.mask_num, clf.thresh, 'clf' , clf.class_score.mean(), average_ns(clf, result='ratio').mean(), average_ns(clf, result='mean').mean()])
        pair_score.append([clf.mask_num, clf.thresh, 'dummy', clf.dummy_score.mean(), 0, 0])
```

```python
pair_score = pd.DataFrame(pair_score, columns=['n_regions', 'threshold', 'clf', 'roc_auc', 'ratio', 'n' ])
pair_score['n_regions'] = pair_score['n_regions'].apply(str)
```

```python
ggplot(pair_score[pair_score.clf=='clf'], aes('threshold', 'roc_auc', group='n_regions', color='n_regions')) + geom_line() + geom_point()# + facet_grid('n_regions')
```

```python
ggplot(pair_score[pair_score.clf=='clf'], aes('threshold', 'ratio', group='n_regions', color='n_regions')) + geom_line() + geom_point()# + facet_grid('n_regions')
```

```python
ggplot(pair_score[pair_score.clf=='clf'], aes('threshold', 'n', group='n_regions', color='n_regions')) + geom_line() + geom_point()# + facet_grid('n_regions')
```

## Discriminability by base rate

```python
from base.tools import region_vox_baserates
```

```python
base_rates = region_vox_baserates(dataset, '../masks/ns_kmeans_all/kmeans_all_20.nii.gz')
p_clf = PairwiseClassifier.load("../results/ns_20_Pairwise_RidgeClassifier_roc_overlap_DM_abs_topics_filt_t_0.08//classifier.pkl")
```

```python
base_rates.columns = ['region', 'base_rate']
acc = pd.DataFrame([np.array(p_clf.class_score.mean(axis=0)), range(1, p_clf.mask_num + 1)]).T
acc.columns = ['auc', 'region']
pl_p = pd.merge(base_rates, acc)
```

```python
%Rpush pl_p
```

```python
%%R
library(ggplot2)
library(dplyr)
library(MASS)


data_p <- cbind(arrange(summarise(group_by(pl_p, region), auc=mean(auc)), auc), factor(1:length(levels(pl_p$region))))
colnames(data_p)[3] <- "order"
plp_p <- left_join(pl_p, data_p, by=c("region"))
```

```python
%%R
data_p <- cbind(arrange(summarise(group_by(pl_p, region), auc=mean(auc)), auc), factor(1:length(levels(pl_p$region))))
colnames(data_p)[3] <- "order"
plp_p <- left_join(pl_p, data_p, by=c("region"))
```

```python
%%R
ggplot(plp_p, aes(order, base_rate, fill=factor(region))) + geom_violin() + theme(legend.position="none")
```

```python
%%R
ggplot(pl_p, aes(auc, base_rate, color=factor(region))) + geom_point(size=0.5, position=position_jitter(width=.0005), alpha=0.7) + 
stat_summary(fun.y=mean, geom="point", size=6) + theme(legend.position="none") 
```

```python
%%R
mean_br_p <- summarise(group_by(pl_p, region), base_rate = mean(base_rate), auc= mean(auc))
```

```python
%%R
print(summary(rlm(auc~base_rate, data=mean_br_p)))
ggplot(mean_br_p, aes(base_rate, auc)) + geom_point() + geom_smooth(method=rlm, alpha=0.5, color='red') 
```

There does seem to be a slight relationship between base rae such that as base rate increases, discriminability decreases. However, it doesn't seem very robust and there are many expections - not driving the effects

```python
n = 60
masks = [(mask, int(mask.split('/')[-1].split('.')[0])) for mask in glob.glob('../masks/ns_kmeans_all/' + str(n) + '/*')]
masks.sort(key=lambda tup: tup[1])
```
