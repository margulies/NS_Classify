```python
plot_score = []
all_scores = []
for clfs in [clfs]:
    for clf in clfs:
        for n, score in enumerate(clf.class_score):
            all_scores.append([clf.mask_num, n, clf.thresh, 'clf' , score, oa.average_ns(clf, mean=False)[n][0]/(oa.average_ns(clf,mean=False)[n][1]*1.0), oa.average_ns(clf, mean=False)[n][0]])
        plot_score.append([clf.mask_num, clf.thresh, 'clf' , clf.class_score.mean(), oa.average_ns(clf)[0]/(oa.average_ns(clf)[1]*1.0), oa.average_ns(clf)[0]])
        plot_score.append([clf.mask_num, clf.thresh, 'dummy', clf.dummy_score.mean(), 0, 0])
```

```python
def get_region_fis(n, comparison = None):
    n = n - 1
    p_fi = pd.DataFrame(p_clf.feature_importances[n])

    # Melt feature importances, and add top_words for each feeature
    p_fi['region'] = range(1, p_fi.shape[0] + 1)
    p_fis_melt = pd.melt(p_fi, var_name='topic_order', value_name='importance', id_vars=['region'])

    top_words = pd.read_csv("../data/unprocessed/abstract_topics_filtered/top_words_80.csv")
    keys = pd.merge(pd.DataFrame(np.array([range(0, 80), p_clf.feature_names]).T, columns=['topic_order', 'topic_name']), top_words)
    keys.topic_order = keys.topic_order.astype('int')

    p_fis_melt= pd.merge(p_fis_melt, keys)

    p_fis_melt = pd.merge(
        p_fis_melt, pd.DataFrame(np.array([p_clf.class_score[n], range(1, p_clf.class_score.shape[0] + 1)]).T, columns=['roc_auc', 'region']))
    p_fis_melt['abs_imp'] = np.abs(p_fis_melt['importance'])

    p_fis_melt =  p_fis_melt[p_fis_melt['region'] != n + 1]
    if comparison is not None:
        p_fis_melt = p_fis_melt[p_fis_melt['region'] == comparison]
        
    p_fis_melt['importance'] = p_fis_melt['importance'] * - 1
    return p_fis_melt
```
