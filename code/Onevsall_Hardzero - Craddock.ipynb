```python
plot_score = []
all_scores = []
for clfs in [clfs_20, clfs_30, clfs_40]:
    for clf in clfs:
        for n, score in enumerate(clf.class_score):
            all_scores.append([clf.mask_num, n, clf.thresh, 'clf' , score, oa.average_ns(clf, mean=False)[n][0]/(oa.average_ns(clf,mean=False)[n][1]*1.0), oa.average_ns(clf, mean=False)[n][0]])
        plot_score.append([clf.mask_num, clf.thresh, 'clf' , clf.class_score.mean(), oa.average_ns(clf)[0]/(oa.average_ns(clf)[1]*1.0), oa.average_ns(clf)[0]])
        plot_score.append([clf.mask_num, clf.thresh, 'dummy', clf.dummy_score.mean(), 0, 0])
```

```python
plot_score = pd.DataFrame(plot_score, columns=['n_regions', 'threshold', 'clf', 'roc_auc', 'ratio', 'n' ])
plot_score['n_regions'] = plot_score['n_regions'].apply(str)

all_scores = pd.DataFrame(all_scores, columns=['n_regions', 'region', 'threshold', 'clf', 'roc_auc', 'ratio', 'n' ])
all_scores['n_regions'] = all_scores['n_regions'].apply(str)
```

# Pairwise analysis

```python
p_20 = [PairwiseClassifier.load(clf  + "/classifier.pkl") for clf in glob.glob("../results/craddock_20_Pairwise_RidgeClassifier_roc_DM_*")]
p_30 = [PairwiseClassifier.load(clf + "/classifier.pkl") for clf in glob.glob("../results/craddock_30_Pairwise_RidgeClassifier_roc_DM_**")]
p_40 = [PairwiseClassifier.load(clf + "/classifier.pkl") for clf in glob.glob("../results/craddock_40_Pairwise_RidgeClassifier_roc_DM_*")]                                                                   
```

```python
pair_score = []
all_scores_pair = []
for clfs in [p_20, p_30, p_40]:
    for clf in clfs:
        for n, score in enumerate(clf.class_score.mean(axis=0)):
            all_scores_pair.append([clf.mask_num, n, clf.thresh, 'clf' , score, pw.average_ns(clf, result='ratio').mean(), pw.average_ns(clf, result='mean').mean()])
        pair_score.append([clf.mask_num, clf.thresh, 'clf' , clf.class_score.mean(), pw.average_ns(clf, result='ratio').mean(), pw.average_ns(clf, result='mean').mean()])
        pair_score.append([clf.mask_num, clf.thresh, 'dummy', clf.dummy_score.mean(), 0, 0])
```
