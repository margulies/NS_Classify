```python
from neurosynth.base.dataset import Dataset
c_20_topics.dataset =  Dataset.load("../data/datasets/dataset_abs_topics_pandas.pkl")
```

```python
ns = []
for i in c_20_topics.c_data.flatten():
    if i is not None:
        X, y = i
        count = np.bincount(y)
        ns.append((count[0]*1.0)/(count[0]+count[1]))
    else:
        ns.append(None)
ns = np.array(ns).reshape(20, 20)
```

```python
ns = np.ma.masked_array(ns)
ns.mask = np.equal(ns, None)
```
