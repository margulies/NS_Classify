## Exploring different thresholds to see how it affects share of junk vs non junk.

Hypothesis was that higher thresholds = higher N = junk more important

```python
clf_01 = cPickle.load(open("../results/ns_k_20_GB_topics_t_0.01/classifier.pkl", 'rb'))
clf_03 = cPickle.load(open("../results/ns_k_20_GB_topics_t_0.03/classifier.pkl", 'rb'))
clf_07 = cPickle.load(open("../results/ns_k_20_GB_topics_t_0.07/classifier.pkl", 'rb'))
clf_1 = cPickle.load(open("../results/ns_k_20_GB_topics_t_0.1/classifier.pkl", 'rb'))
```

```python
for clf in [clf_01, clf_03, clf_07, clf_1]:
    
    print
    print clf.thresh
    calculate_feature_corr(clf)
    fc_mean = clf.feature_corr.mean(axis=0).T
    heat_map(fc_mean, np.array(range(0, clf.mask_num))+1, clf.feature_names)
    
    junk_fi_pearson = clf.feature_importances[:, :, ix_subset(dataset_topics_40, junk_topics_2)]
    cog_fi_pearson = clf.feature_importances[:, :, ix_subset(dataset_topics_40, reduced_topics_2)]
    
    print "Junk Topics"
    print "Mean:", junk_fi_pearson.mean()
    print "Var:", junk_fi_pearson.var()

    print
    print "Cognitive Topics"
    print "Mean:", cog_fi_pearson.mean()
    print "Var:", cog_fi_pearson.var()
```

Results are actually slightly opposite of expected. Greater threshold result in (slightly) more balance in the feature impotances between junk and cognitive. Perhaps because the problem becomes harder?

This is consitent with cognitive topics being used more later in the trees - for difficult cases
